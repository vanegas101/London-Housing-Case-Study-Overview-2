```python
url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"
```

```python
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col=None)
```

```python
properties.head()
```

```python
#transposed the columns
properties = properties.transpose()
```

```python
properties.reset_index().head()
```

```python
properties.columns
```

```python
# copied the dataframe so I could make changes to it without compromising the actual dataframe

test2 = properties.copy()
```

```python
test2.head()
```

```python
#reset index to add the index back
test2.reset_index().head()
```

```python
#Replaced the columns with the first row
test2.columns = test2.iloc[0]
```

```python
test2.head()
```

```python
#dropped the first row 
test2 = test2.drop(test2.index[0])
```

```python
test2.head()
```

```python
#dropped the column labeled NaT
test2 = test2.drop(columns='NaT')
```

```python
test2.head()
```

```python
#added the index back
test2.reset_index(inplace = True)
```

```python
#changed some of the column names
test2.rename(columns={'index': 'Boroughs'}, inplace=True)
```

```python
test2.columns.names = ['index']
```

```python
#drop the city of london row
test2.drop(test2.index[0])
```

```python
test2.shape
```

```python
#turned prices into floats
test2 = test2.apply(pd.to_numeric, errors='ignore')
```

```python
#plotted the dataframe to show the increase in prices for each borough over time
plt.figure(figsize=(12, 8))
for borough in test2['Boroughs'].unique():
    plt.plot(test2.columns[1:], test2[test2['Boroughs'] == borough].iloc[0, 1:], label=borough)

plt.title('Average House Prices in London Boroughs Over Time')
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()

```

```python
#Calculated the ratio of the Average Price for 1998 divided by the Average Price for 2018 
#for each borough

def create_price_ratio(df):
    avg_price_1998 = test2['1998-01-01'].mean()
    avg_price_2018 = test2['2018-01-01'].mean()
    
    ratio = avg_price_1998 / avg_price_2018
    
    return ratio

ratios = {}
for borough in test2['Boroughs'].unique():
    
    borough_df = test2[test2['Boroughs'] == borough]
    
    ratio = create_price_ratio(borough_df)
    
    ratios[borough] = ratio

for borough, ratio in ratios.items():
    print(f'{borough}: {ratio}')
```
